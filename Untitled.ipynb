## import

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from tqdm import tqdm

import matplotlib.pyplot as plt
import numpy as np

import torchvision
from torchvision import datasets
from torchvision import transforms

from torchcp.classification.predictors import SplitPredictor, ClusterPredictor, ClassWisePredictor, WeightedPredictor
from torchcp.classification.scores import THR, APS, SAPS, RAPS, Margin
```

## prepare

```python
BATCH_SIZE = 128
LEARNING_RATE = 1e-1
MOMENTUM = 0.9
STEP=5
GAMMA=0.5
NUM_EPOCHS = 30
dropout_p = 0.2
alpha = 0.1

criterion = nn.CrossEntropyLoss()
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(torch.cuda.is_available())
```

## dataset

```python
def load_mnist():
    transform = transforms.Compose([transforms.ToTensor(),
                                    transforms.Normalize((0.5,), (0.5,))])

    train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
    test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

    train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
    test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)
    return train_dataloader, test_dataloader
```

```python
def load_cifar10():
    transform = transforms.Compose([transforms.ToTensor(),
                                    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))])

    train_set = torchvision.datasets.CIFAR10(root='../data', train=True,download=True, transform=transform)
    test_set = torchvision.datasets.CIFAR10(root='../data', train=False,download=True, transform=transform)

    train_dataloader = torch.utils.data.DataLoader(train_set, batch_size=128, shuffle=True, num_workers=2)
    test_dataloader = torch.utils.data.DataLoader(test_set, batch_size=128,shuffle=False, num_workers=2)
    return train_dataloader, test_dataloader
```

## model

```python
class ConvNetCifar(nn.Module):
    def __init__(self):
        super(ConvNetCifar, self).__init__()
        self.conv1 = nn.Conv2d(3, 4, 3)  
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(4, 8, 3)
        self.fc1 = nn.Linear(8 * 6 * 6, 32)
        self.fc2 = nn.Linear(32, 10)#分类到多少class

    def forward(self, x):
        x = self.pool(F.leaky_relu(self.conv1(x), negative_slope=0.01))
        x = self.pool(F.leaky_relu(self.conv2(x), negative_slope=0.01))
        x = x.view(-1, 8 * 6 * 6)
        x = F.leaky_relu(self.fc1(x), negative_slope=0.01)
        x = self.fc2(x)
        return x
```

```python
class ConvNetMnist(nn.Module):
    def __init__(self):
        super(ConvNetMnist, self).__init__()
        # 修改第一个卷积层以接受单通道输入
        self.conv1 = nn.Conv2d(1, 4, 3, padding=1)  
        self.pool = nn.MaxPool2d(2, 2)
        # 输出通道保持不变，调整内核大小和填充以保持尺寸
        self.conv2 = nn.Conv2d(4, 8, 3, padding=1)
        # 修改全连接层以匹配经过两次池化的 7x7 图像
        self.fc1 = nn.Linear(8 * 7 * 7, 32)
        # MNIST 有 10 个类，所以最后一层保持不变
        self.fc2 = nn.Linear(32, 10)

    def forward(self, x):
        x = self.pool(F.leaky_relu(self.conv1(x), negative_slope=0.01))
        x = self.pool(F.leaky_relu(self.conv2(x), negative_slope=0.01))
        # 修改view调用以匹配新的维度大小
        x = x.view(-1, 8 * 7 * 7)
        x = F.leaky_relu(self.fc1(x), negative_slope=0.01)
        x = self.fc2(x)
        return x
```

## train

```python
def training(train_dataloader, test_dataloader,NUM_EPOCHS,dataset):
    
    if dataset == "mnist":
        model = ConvNetMnist()
    else:
        model = ConvNetCifar()
    model.to(device)
    optimizer = optim.SGD(model.parameters(), lr=LEARNING_RATE, momentum=MOMENTUM)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=STEP, gamma=GAMMA)
    
    
    training_loss = []
    for epoch in tqdm(range(NUM_EPOCHS), desc="Training Progress"):
        model.train()
        torch.cuda.empty_cache()

        running_cls_loss = 0.0
        for batch_idx, (image, target) in enumerate(train_dataloader):
        
            ## 计算loss
            image = image.to(device)
            target = target.to(device)
            output = model(image)
            loss = criterion(output,target)
            running_cls_loss += loss.item()
            loss.backward()
        
        ## optimizer
            optimizer.step()
            optimizer.zero_grad()

        epoch_loss = running_cls_loss / len(train_dataloader)
        training_loss.append(epoch_loss)
        # change learning rate
        scheduler.step()
    if model is None:
        raise ValueError("Model was not returned from the training function")

        
    return model
```

## evaluation

```python
def plot_result(coverage,average,score_name,dataset):
    predictors = ["split","cluster","classwise"]
    # 设置柱状图的位置和宽度
    x = np.arange(len(predictors))  # 预测器标签的位置
    width = 0.35  # 柱状图的宽度

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, average, width, label='Average Size')
    rects2 = ax.bar(x + width/2, coverage, width, label='Coverage Rate')

    # 添加标签、标题和自定义X轴刻度标签等
    ax.set_ylabel('Scores')
    ax.set_title(f'Scores by predictor: {score_name} on dataset {dataset}')
    ax.set_xticks(x)
    ax.set_xticklabels(predictors)
    ax.legend()

    # 绘制数值标签
    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)

    # 显示图表
    plt.show()
    
```

```python
def evaluation_model(model,alpha,train_dataloader,test_dataloader,dataset):
    scores = [THR(),  APS(), RAPS(1, 0), SAPS(0.2), Margin()]
    predictors = ["split","cluster","classwise"]

    for score in scores:
        coverage = []
        average = []
        for name in predictors:
            if name == "split": predictor = SplitPredictor(score, model)
            elif name == "cluster": predictor =  ClusterPredictor(score, model)
            elif name == "classwise": predictor =  ClassWisePredictor(score, model)
            
            predictor.calibrate(train_dataloader, alpha)
            evaluation_results = predictor.evaluate(test_dataloader)
            coverage.append(evaluation_results['Coverage_rate'])
            average.append(evaluation_results['Average_size'])
        print(coverage,average)
        
        plot_result(coverage,average,score.__class__.__name__,dataset)
```

## plot result

```python
m_train_dataloader, m_test_dataloader = load_mnist()
model_mnist = training(m_train_dataloader, m_test_dataloader,10,"mnist")
evaluation_model(model_mnist,0.1,m_train_dataloader, m_test_dataloader,"mnist")
```

```python
c_train_dataloader, c_test_dataloader = load_cifar10()
model_cifar = training(c_train_dataloader, c_test_dataloader,30,"cifar10")
evaluation_model(model_cifar,0.1,c_train_dataloader, c_test_dataloader,"cifar10")
```

## find best weight

```python
def plot_best(coverage,average,predictor,dataset,option):
    weights = [5,10,15,20,25,30]
    x = np.arange(len(weights))
    width = 0.35
    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, average, width, label='Average Size')
    rects2 = ax.bar(x + width/2, coverage, width, label='Coverage Rate')
    ax.set_ylabel('Scores')
    if option == "SAPS":
        ax.set_title(f'Scores by predictor: SAPS with predictor:{predictor} on dataset {dataset}')
    else:
        ax.set_title(f'Scores by predictor: RAPS with predictor:{predictor} on dataset {dataset}')
    ax.set_xticks(x)
    ax.set_xticklabels(weights)
    ax.legend()
    
    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)

    # 显示图表
    plt.show()
    
    
```

```python
def find_best_weight(model,alpha,train_dataloader, test_dataloader,dataset):
    weights = [5,10,15,20,25,30]
    predictors = ["split","cluster","classwise"]
    for name in predictors:
        coverage = []
        average = []
        
        for weight in weights:
            score = SAPS(weight = weight)
            if name == "split": predictor = SplitPredictor(score, model)
            elif name == "cluster": predictor =  ClusterPredictor(score, model)
            elif name == "classwise": predictor =  ClassWisePredictor(score, model)
                
            predictor.calibrate(train_dataloader, alpha)
            evaluation_results = predictor.evaluate(test_dataloader)
            coverage.append(evaluation_results['Coverage_rate'])
            average.append(evaluation_results['Average_size'])
        print(coverage,average)
        
        plot_best(coverage,average,predictor.__class__.__name__,dataset,"SPAS")
```

```python
find_best_weight(model_mnist,0.1,m_train_dataloader, m_test_dataloader,"mnist")
```

```python
find_best_weight(model_cifar,0.1,c_train_dataloader, c_test_dataloader,"cifar")
```

```python
def find_best_penalty(model,alpha,train_dataloader, test_dataloader,dataset):
    penaltys = [5,10,15,20,25,30]
    predictors = ["split","cluster","classwise"]
    for name in predictors:
        coverage = []
        average = []
        
        for penalty in penaltys:
            score = RAPS(penalty = penalty)
            if name == "split": predictor = SplitPredictor(score, model)
            elif name == "cluster": predictor =  ClusterPredictor(score, model)
            elif name == "classwise": predictor =  ClassWisePredictor(score, model)
                
            predictor.calibrate(train_dataloader, alpha)
            evaluation_results = predictor.evaluate(test_dataloader)
            coverage.append(evaluation_results['Coverage_rate'])
            average.append(evaluation_results['Average_size'])
        print(coverage,average)
        
        plot_best(coverage,average,predictor.__class__.__name__,dataset,"RAPS")
```

```python
find_best_penalty(model_mnist,0.1,m_train_dataloader, m_test_dataloader,"mnist")
```
